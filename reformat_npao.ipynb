```python
# check list of column, to reduce columns
npao_original.columns
```

```python
npao_reduced = npao_original[['YearStart', 'LocationDesc', 'GeoLocation', 'Class', 
                              'Question', 'Data_Value', 'Data_Value_Alt', 'Sample_Size']]
npao_reduced.head()
```

```python
# checking if Data value and data value alt are the same throughout
npao_reduced["value_same_alt"] = npao_reduced["Data_Value"] == npao_reduced["Data_Value_Alt"]  
npao_reduced["value_same_alt"].value_counts()
```

```python
npao_reduced.loc[npao_reduced["value_same_alt"] == False, :].head()
```

```python
npao_reduced = npao_reduced.dropna(how='any')
npao_reduced.info()
```

```python
# removing columns that are not needed
npao_reduced = npao_reduced[['LocationDesc', 'GeoLocation', 'Class', 'Question', 'Data_Value', 'Sample_Size']]
npao_reduced.tail()
```

```python
# need to create a new column- the product of sample size times data value (to later calculate weighted average)
npao_reduced["ValueXSample"] = npao_reduced["Data_Value"] * npao_reduced["Sample_Size"]
npao_reduced.head()
```

```python
# reformatting the data frame
npao = npao_reduced.groupby(["LocationDesc", "GeoLocation", "Class", "Question"])["ValueXSample", "Sample_Size"].sum()
npao
```

```python
# converting the groupby object to a dataframe
npao = npao.reset_index()
npao
```

```python
# calculating a column with the weighted average rate per question
npao["Average %"] = npao["ValueXSample"] / npao["Sample_Size"]
npao
```

```python
# need to recreate the data frame such that each question would appear in a different column 
#and will have it's corresponding average %

# first, get rid of columns that are not needed
npao = npao[["LocationDesc", "GeoLocation", "Question", "Average %"]]
npao
```

```python
# start with obesity
obesity_df = npao.loc[npao["Question"] == 'Percent of adults aged 18 years and older who have obesity', :].reset_index(drop=True)
obesity_df
```

```python
# grab the columns as arrays
locations = obesity_df["LocationDesc"]
geo = obesity_df["GeoLocation"]
obesity = obesity_df["Average %"]
```

```python
# getting all other vars
overweight_df = npao.loc[npao["Question"] == 'Percent of adults aged 18 years and older who have an overweight classification', :].reset_index(drop=True)
inactivity_df = npao.loc[npao["Question"] == 'Percent of adults who engage in no leisure-time physical activity', :].reset_index(drop=True)
strength_df = npao.loc[npao["Question"] == 'Percent of adults who engage in muscle-strengthening activities on 2 or more days a week', :].reset_index(drop=True)                                                                                                                                                                      
conditioning_low_df = npao.loc[npao["Question"] == 'Percent of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity or 75 minutes a week of vigorous-intensity aerobic activity (or an equivalent combination)', :].reset_index(drop=True)                                                                                                                                                                                     
conditioning_high_df = npao.loc[npao["Question"] == 'Percent of adults who achieve at least 300 minutes a week of moderate-intensity aerobic physical activity or 150 minutes a week of vigorous-intensity aerobic activity (or an equivalent combination)', :].reset_index(drop=True)                                                                                                                                                                                      
strength_conditioning_df = npao.loc[npao["Question"] == 'Percent of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity or 75 minutes a week of vigorous-intensity aerobic physical activity and engage in muscle-strengthening activities on 2 or more days a week', :].reset_index(drop=True)                                                                                                                                                              
no_veggies_df = npao.loc[npao["Question"] == 'Percent of adults who report consuming vegetables less than one time daily', :].reset_index(drop=True)                                                  
no_fruit_df = npao.loc[npao["Question"] == 'Percent of adults who report consuming fruit less than one time daily', :].reset_index(drop=True)                                                   
no_fruit_df
```

```python
# building the new data frame
npao_final = pd.DataFrame({"State": locations, "Long & Lat":geo, "Obesity %": obesity, 
                           "Overweight %": overweight_df["Average %"], 
                           "In-Active %": inactivity_df["Average %"], 
                           "Muscle Strengthening %": strength_df["Average %"], 
                           "Low Aerobics %": conditioning_low_df["Average %"], 
                           "High Aerobics %": conditioning_high_df["Average %"], 
                           "Strength & Conditioning %": strength_conditioning_df["Average %"], 
                           "Low Veggies Consumption %": no_veggies_df["Average %"], 
                           "Low Fruit Consumption %": no_fruit_df["Average %"]})
npao_final.head()
```

```python
npao_final.to_csv("Datasets/npao_final.csv", encoding="utf-8", index=False, header=True)
```
